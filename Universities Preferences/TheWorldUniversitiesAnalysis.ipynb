```python
data = df.copy()
#Converting the International Students from character  to numeric values
data["International Students"] = data["International Students"].apply(lambda x : pd.to_numeric(x.strip("%")))
```

```python
data["Number International Students"] = np.round(data["Student Population"] * (data["International Students"]/100))
```

```python
data.dropna(inplace=True)
```

```python
data[['Female Students Ratio1', 'Male Students Ratio2', "Error"]] = data['Female to Male Ratio'].str.split(':', expand=True)
data['Female Students Ratio1'] = data['Female Students Ratio1'].astype(float)
data['Male Students Ratio2'] = data['Male Students Ratio2'].astype(float)
data["Female Students"] = ((data["Female Students Ratio1"]*data["Student Population"])/(data["Male Students Ratio2"]+data["Female Students Ratio1"])).round()
data["Male Students"] = data["Student Population"] - data["Female Students"]
data = data.drop(["Error","Female Students Ratio1",'Male Students Ratio2','International Students'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data.loc[data["Female Students"].isna(), "Female Students"] = pd.to_numeric(data.loc[data["Female Students"].isna(), "Female to Male Ratio"]) * data.loc[data["Female Students"].isna(), "Student Population"]
data.loc[data["Male Students"].isna(), "Male Students"] = data.loc[data["Male Students"].isna(), "Student Population"] - data.loc[data["Male Students"].isna(), "Female Students"]
```

```python
data = data.drop("Female to Male Ratio",axis = 1)
```

```python
# Define a custom color palette
custom_palette = sns.color_palette("flare", 14)

# Copy the data and drop the 'Year' column
violin_data = data.copy()
violin_data = violin_data.drop("Year", axis=1)

# Standardize the data
for col in violin_data.columns[3:]:
    mu = violin_data[col].mean()
    sigma = violin_data[col].std()
    violin_data[col] = (violin_data[col] - mu) / sigma

# Plotting violin plots
plt.figure(figsize=(12, 8))
sns.violinplot(data=violin_data.iloc[:, 3:], palette=custom_palette, inner="quartile")
plt.title('Distribution of Standardized Scores')
plt.xlabel('Metrics')
plt.ylabel('Standardized Scores')
plt.xticks(rotation=90)
plt.show()
```

```python
# Count the number of universities per country
universities_per_country = data['Country'].value_counts()

# Get the top 10 and bottom 10 countries with the most universities
top_countries = universities_per_country.head(10)
bottom_countries = universities_per_country.tail(10)

# Plotting the top 10 countries
plt.figure(figsize=(12, 8))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='plasma')
plt.title('Top 10 Countries with the Most Universities')
plt.xlabel('Number of Universities')
plt.ylabel('Country')
plt.xticks(rotation=45)
plt.show()

# Plotting the bottom 10 countries
plt.figure(figsize=(12, 8))
sns.barplot(x=bottom_countries.values, y=bottom_countries.index, palette='rainbow')
plt.title('Bottom 10 Countries with the Fewest Universities')
plt.xlabel('Number of Universities')
plt.ylabel('Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Get the top 5 countries with the highest overall scores
top_countries_overall = data.groupby('Country')['Overall Score'].mean().nlargest(5).index

# Filter the data for these top 5 countries
top_countries_data = data[data['Country'].isin(top_countries_overall)]

# Calculate the mean student population, industry impact, and overall score for each country
country_stats = top_countries_data.groupby('Country').agg({'Student Population':'mean', 
                                                           'Industry Impact':'mean',
                                                           'Overall Score':'mean',
                                                           'Research Quality': 'mean',
                                                           'Research Environment': 'mean'}).reset_index()

# Sort the country_stats DataFrame by Overall Score in descending order
country_stats = country_stats.sort_values(by='Overall Score', ascending=False)

# Create subplots
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plotting the comparison for mean student population
sns.barplot(data=country_stats, x='Student Population', y='Country', palette='viridis', ax=ax1)
ax1.set_title('Mean Student Population and Industry Impact for Top 5 Countries with Highest Overall Scores')
ax1.set_xlabel('Mean Student Population')
ax1.set_ylabel('Country')

# Create a second y-axis for mean industry impact
ax2 = ax1.twiny()
sns.barplot(data=country_stats, x='Industry Impact', y='Country', palette='plasma', ax=ax2, alpha=0.5)
ax2.set_xlabel('Mean Industry Impact')

plt.show()

# Plotting the comparison for mean research quality
plt.figure(figsize=(12, 8))
sns.barplot(data=country_stats, x='Research Quality', y='Country', palette='inferno')
plt.title('Mean Research Quality for Top 5 Countries with Highest Overall Scores')
plt.xlabel('Mean Research Quality')
plt.ylabel('Country')
plt.show()

# Plotting the comparison for mean overall score
plt.figure(figsize=(12, 8))
sns.barplot(data=country_stats, x='Overall Score', y='Country', palette='magma')
plt.title('Mean Overall Score for Top 5 Countries with Highest Overall Scores')
plt.xlabel('Mean Overall Score')
plt.ylabel('Country')
plt.show()

# Plotting the comparison for mean research environment
plt.figure(figsize=(12, 8))
sns.barplot(data=country_stats, x='Research Environment', y='Country', palette='autumn')
plt.title('Mean Research Environment for Top 5 Countries with Highest Overall Scores')
plt.xlabel('Mean Research Environment')
plt.ylabel('Country')
plt.show()
```
