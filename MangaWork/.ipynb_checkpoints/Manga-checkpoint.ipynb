```python
import pandas as pd
```

```python
# Load the best-selling manga dataset
manga_dataset = pd.read_csv('best-selling-manga.csv')

# Load the Anime Top 10000 dataset
anime_dataset = pd.read_csv('Anime_Top10000.csv')
```

```python
# Display the first few rows of the manga dataset
print(manga_dataset.head())

# Display the first few rows of the anime dataset
print(anime_dataset.head())
```

```python
# Check the shape of the manga dataset
print(manga_dataset.shape)

# Check the shape of the anime dataset
print(anime_dataset.shape)

```

```python
# View the columns of the manga dataset
print(manga_dataset.columns)

# View the columns of the anime dataset
print(anime_dataset.columns)
```

```python
# Summary statistics for the manga dataset
print(manga_dataset.describe())

# Summary statistics for the anime dataset
print(anime_dataset.describe())
```

```python
# Check for missing values in the manga dataset
print(manga_dataset.isnull().sum())

# Check for missing values in the anime dataset
print(anime_dataset.isnull().sum())
```

```python
# Sort the dataset by 'Approximate sales in million(s)' in descending order and select top 10
top_manga = manga_dataset.sort_values(by='Approximate sales in million(s)', ascending=False).head(10)
print(top_manga[['Manga series', 'Approximate sales in million(s)']])
```

```python
# Group by 'Publisher' and calculate the average of 'Average sales per volume in million(s)'
avg_sales_by_publisher = manga_dataset.groupby('Publisher')['Average sales per volume in million(s)'].mean().sort_values(ascending=False)
print(avg_sales_by_publisher)
```

```python
import matplotlib.pyplot as plt

# Group by 'Demographic' and calculate the total sales
sales_by_demographic = manga_dataset.groupby('Demographic')['Approximate sales in million(s)'].sum()

# Plot a bar chart
sales_by_demographic.plot(kind='bar', figsize=(10, 6))
plt.title('Total Sales by Demographic')
plt.xlabel('Demographic')
plt.ylabel('Approximate Sales in Millions')
plt.show()
```

```python
# Sort the dataset by 'Anime_Rating' in descending order and select top 10
top_anime = anime_dataset.sort_values(by='Anime_Rating', ascending=False).head(10)
print(top_anime[['Anime_Name', 'Anime_Rating']])
```

```python
# Group by 'Anime_Air_Years' and calculate the average rating
avg_rating_by_years = anime_dataset.groupby('Anime_Air_Years')['Anime_Rating'].mean().sort_index()

# Plot a line chart
avg_rating_by_years.plot(kind='line', figsize=(12, 6), marker='o')
plt.title('Average Rating by Air Years')
plt.xlabel('Air Years')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```
